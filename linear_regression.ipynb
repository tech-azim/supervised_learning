```python
import numpy as np
import matplotlib.pyplot as plt

# ===============================
# 1. Data
# ===============================
X = np.array([1, 2, 3, 4, 5])
Y = np.array([2, 3, 5, 6, 8])

# ===============================
# 2. Hitung parameter regresi
# ===============================
X_mean = np.mean(X)
Y_mean = np.mean(Y)

numerator = np.sum((X - X_mean) * (Y - Y_mean))
denominator = np.sum((X - X_mean) ** 2)

m = numerator / denominator
b = Y_mean - m * X_mean

print("Slope (m):", m)
print("Intercept (b):", b)

# ===============================
# 3. Prediksi
# ===============================
Y_pred = m * X + b

# ===============================
# 4. Error / Residual
# ===============================
error = Y - Y_pred
print("\nResidual:")
print(error)

# ===============================
# 5. MSE & RMSE
# ===============================
mse = np.mean(error ** 2)
rmse = np.sqrt(mse)

print("\nMSE:", mse)
print("RMSE:", rmse)

# ===============================
# 6. R-squared (R²)
# ===============================
ss_total = np.sum((Y - Y_mean) ** 2)
ss_residual = np.sum(error ** 2)
r2 = 1 - (ss_residual / ss_total)

print("R²:", r2)

# ===============================
# 7. Visualisasi
# ===============================
plt.scatter(X, Y, label='Data Asli')
plt.plot(X, Y_pred, label='Garis Regresi')

# Garis error (residual)
for i in range(len(X)):
    plt.plot([X[i], X[i]], [Y[i], Y_pred[i]], linestyle='--')

plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.title('Linear Regression dengan Residual')
plt.show()
```
